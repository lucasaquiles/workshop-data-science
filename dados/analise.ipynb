```python
adults = adults.merge(countries, left_on=13, right_on='nome')
adults.head()
```

```python
chaves = ['continentes', '<=50K	']
adults[chaves].groupby('continentes').sum().head()
#agrupa os dados
```

```python
adults['>=50k'] = adults[14] == '>=50k'
adults['<=50k'] = adults[14] == '>50k'
adults.head()
```

```python
from sklearn import svm
clf = svm.SVC(gamma='scale')
clf.fit(adults[[0,2]], adults[14])
```

```python
adults_tests = pd.read_csv('adult.test', header=None, engine="python", sep=", ")
```

```python
clf.score(adults_tests[[0,2]], adults_tests[14])
```
